```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
edge_list_path = './new_graph.txt'
```

# Load The graph from the edge list text file

```python
G = nx.read_edgelist('./new_graph.txt',create_using = nx.Graph(),nodetype = str)
```

# Summary of the Graph

```python
print(nx.info(G))
```

# Functions and attributes in Graph object G (for reference)

```python
help(G)
```

# Plotting the graph

```python
layout_spring = nx.spring_layout(G,k=0.75,iterations=100)
```

```python
layout_kamada = nx.kamada_kawai_layout(G)
```

```python
nx.draw_networkx(G,node_size=200,k=10,iterations=100)
```

# Get connected components

```python
list(nx.connected_components(G))
```

# Betweeness centrality

```python
betweeness_centrality = nx.betweenness_centrality(G)
```

```python
for x in sorted(betweeness_centrality,key=lambda x: betweeness_centrality[x],reverse=True):
    print(x,betweeness_centrality[x])
```
